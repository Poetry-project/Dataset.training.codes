### Remove any duplication.

#### clacuate row is duplacte

```python
if df1.duplicated().sum() > 0 :
    print('Number of duplicated row:', df1.duplicated().sum())
    df1 = df1.drop_duplicates(inplace = True)
    print('Number of duplicated row ِafter remove duplicated:', df1.duplicated().sum())
```

#### count both ‘NaN’ and ‘null’ values

```python
nan_count = df1.isna().sum().sum()
null_count = df1.isnull().sum().sum()
total_missing = nan_count + null_count

print('Number of missing values:', total_missing)
print ("Count of poem text is null={}".format(df1.poem_text.isnull().sum()))
print ("Present poem text is null in dataset={:2.1f}%".format(df1.poem_text.isnull().sum()/len(df1) * 100))
print ("Count of poem text is nan={}".format(df1.poem_text.isna().sum()))
print ("Present poem text is nan in dataset={:2.1f}%".format(df1.poem_text.isna().sum()/len(df1) * 100))
```

```python
#dfresult = df.dropna(how='all', axis=1)
df1 = df1.dropna(how='any',axis=0)
```

```python
#test if any null or nan after remove
nan_count = df1.isna().sum().sum()
null_count = df1.isnull().sum().sum()
total_missing = nan_count + null_count
print('Number of missing values:', total_missing)
```

```python
# Select unique values from feuture category
df1['category'].unique()
```

```python
list_country = ['الإمارات', 'البحرين', 'الجزائر', 'السعودية', 'السودان', 'العراق','المغرب'
                , 'اليمن' , 'تونس', 'سوريا', 'عمان', 'فلسطين', 'لبنان'
                ,'ليبيا', 'مصر', 'الأردن', 'الكويت', 'قطر', 'موريتانيا' ]

list_ear = ['العصر الجاهلي', 'العصر الإسلامي', 'العصر العباسي','العصر الايوبي'
            , 'العصر العثماني', 'عصر المخضرمون', 'العصر الاموي'
            ,'العصر الأندلسي', 'العصر المملوكي']

def change_country_era(text):
    if text in list_country :
        return 'العصر الحديث'
    return text

def change_country_era_to_null(text):
    if text not in list_ear :
        return text
```

```python
#  change country to era in dataset 1
df1['era'] = df1.category.apply(lambda x: change_country_era(x))
```

```python
#  change country to era in dataset 2
df1['country'] = df1.category.apply(lambda x: change_country_era_to_null(x))
```

```python
df1.head()
```

```python
#clacuate count null in feuture country after replace
print ("Count of country is null= {}".format(df1.country.isnull().sum()))
print ("Present country is null in dataset= {:2.1f}%".format(df1.country.isnull().sum()\
                                                             /len(df1) * 100))
```

```python
#Rename fueture name
df1 = df1.rename(columns={"id": "poem_id"})
df1 = df1.rename(columns={"category": "poet_cat"})
```

```python
#Remove feutuer don't use it
df1 = df1.drop('poet_cat', axis=1)
```

```python
df1.columns
```

```python
df1.to_csv("process_data/df1_Add_era_country.csv", encoding='utf-8-sig', index=False , header=True )
```
