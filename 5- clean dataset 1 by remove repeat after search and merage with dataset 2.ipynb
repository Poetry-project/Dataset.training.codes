## Read poem_id as list for compaer in dataset 2

```python
df_repeat_id_list = df_repeat.poem_id.to_list()
print(len(df_repeat_id_list))
#print (df_repeat_id_list)
```

```python
df1.shape[0]
```

```python
#create new dataframe for save a new dataset
df_append = pd.DataFrame(columns=df1.columns)#strict only not repeat in dataset1

for index, row in tqdm(df1.iterrows(), total=df1.shape[0]):
    if df1['poem_id'].loc[index] not in df_repeat_id_list :
        df_append.loc[len(df_append)] = df1.loc[index]

print(df_append.shape)
df_append.head()
```

# Save Dataset1 to new file after clean from repeat row

```python
df_append.to_csv("process_data/Arabic_poetry_dataset_clean.csv", index=False)
```
